# Bresenham

```python
start_point = [0,0]
end_point = [10,16]

#----- INITIALIZATION ----------------------------------
x1 = start_point[0]
y1 = start_point[1]
x2 = end_point[0]
y2 = end_point[1]

#----- CALCULATE M ------------------------------------------
m = (y2-y1)/(x2-x1)

#----- CALCULATE DELTA --------------------------------
dx = x2 - x1
dy = y2 - y1
if m > 0 and m < 1:
    d1 = 2 *dy
    d2 = 2 * (dx - dy)
    p = d1 - dx
else:
    d1 = 2 *dx
    d2 = 2 * (dx - dy)
    p = d1 - dy

#----- CALCULATE INCREMENT X AND P --------------------------
x_list = []
p_list = []
y_list = []

def increment_x_1(x1, x2):
    if x1 < x2:
        #print(x1)
        x_list.append(x1)
        return increment_x_1(x1 + 1, x2)
    else:
        x_list.append(x2)
    
    return x_list

def increment_x_2(x1, x2, p):
    if x1 < x2:
        if p >= 0:
            x_list.append(x1)
            p_list.append(p)
            return increment_x_2(x1+1, x2, p + d2)
        else :  
            x_list.append(x1)
            p_list.append(p)
            return increment_x_2(x1, x2, p + d1)
      
    else:
        x_list.append(x1)
        p_list.append(p)
        
    return x_list, p_list 

#----- CALCULATE INCREMENT Y AND P --------------------------

def increment_y_1(y1, y2, p):
    if y1 < y2:
        if p >= 0:
            y_list.append(y1)
            p_list.append(p)
            return increment_y_1(y1+1, y2, p - d2)
        else :  
            y_list.append(y1)
            p_list.append(p)
            return increment_y_1(y1, y2, p + d1)
      
    else:
        y_list.append(y1)
        y_list.append(y2)
        p_list.append(p)
        p_list.append(p_list[-2])
        
    return y_list, p_list

def increment_y_2(y1, y2):
    if y1 < y2:
        y_list.append(y1)
        return increment_y_2(y1 + 1, y2)
    else : 
        y_list.append(y2)
    
    return y_list

if m > 0 and m < 1:
    increment_x_1(x1,x2)
    increment_y_1(y1,y2,p)
else:
    increment_x_2(x1,x2,p)
    increment_y_2(y1,y2)


def display():
    
    print(
    f"""
    ===============================================================
                INITIALIZATION OF START POINT AND END POINT
    ===============================================================
    start point = {start_point}
      end point = {end_point}
    
    ===============================================================
                INITIALIZATION OF X AND Y
    ===============================================================
    x1 = {x1} | x2 = {x2}
    y1 = {y1} | y2 = {y2}
    
    ===============================================================
                GET M
    ===============================================================
    m = ({y2}-{y1})/({x2}-{x1}) = {(y2-y1)/(x2-x1)}
    
    ===============================================================
                GET DELTA 
    ===============================================================
    dx : {x2} - {x1} = {x2 - x1}
    dy : {y2} - {y1} = {y2 - y1}
    
    d1 : {d1}
    d2 : {d2}
    
    p  : {p}
    ===============================================================
                INCREMENT 
    ===============================================================
    p = {p_list}
    x = {x_list}
    y = {y_list}

    """)
```

```python
import matplotlib.pyplot as plt
import numpy as np

display()
plt.figure(figsize=(15,8))
plt.xticks(np.arange(-5, y_list[-1]+1, step=1))
plt.yticks(np.arange(-5, y_list[-1]+1, step=1))
plt.scatter(x_list,y_list, color='r')
plt.plot(x_list,y_list, color= 'b')
plt.grid(True)
plt.title("Primitive Graphics")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
